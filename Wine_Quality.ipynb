<h3> KNN Modelling - 1 </h3>
<h4>  StandardScaler() and LabelEncoder() are used to standardise and scale the data.<h4> 
  

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

X=red_wines.drop(['quality','free sulfur dioxide'],axis=1).values
scaler=StandardScaler()

Y=red_wines['quality'].values
labelencoder =LabelEncoder()
Y= labelencoder.fit_transform(Y)

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42,stratify=Y)

scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
neighbors = np.arange(1, 60)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

for i, k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,Y_train)
    train_accuracy[i] = knn.score(X_train, Y_train)
    test_accuracy[i] = knn.score(X_test, Y_test)

plt.title('Accuracy for K values')
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
error_rate = []

for i in range(1,60):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, Y_train)
    pred = knn.predict(X_test)
    error_rate.append(np.mean(pred != Y_test))

plt.figure(figsize=(10,5))
plt.title('Error Rates for K values')
plt.plot(range(1,60),error_rate, marker='o', markersize=7)
plt.show()
```

<h4> A small value of K leads to overfitting and a large value of K leads to underfitting. From the accuracy and error graphs, K=12 is an optimal value. </h4>

```python
knn=KNeighborsClassifier(n_neighbors=12)
knn.fit(X_train,Y_train)
Y_pred=knn.predict(X_test)
print('K = 12')
print('Accuracy Score = ',accuracy_score(Y_test,Y_pred))
```

<h4> Using a basic modelling technique and K value of 12, Accuracy = 0.59375 </h4>


<h3> KNN Modelling - 2 </h3>
<h4> Outliers are removed and then modelling is performed. </h4>

```python
data=red_wines.copy()
#using red_wines dataframe will cause inplace removal in red_wines

for feature in data:
    if (feature=='quality') : continue
        
    Q1 = np.percentile(data[feature], q=25)
    Q3 = np.percentile(data[feature], q=75)
    interquartile_range = Q3 - Q1
    step = 1.5 * interquartile_range
    
    dropindex = data[~((data[feature] >= Q1 - step) & (data[feature] <= Q3 + step))].index
    data.drop(dropindex,inplace=True)
```

```python
x=data.drop(['quality','free sulfur dioxide'],axis=1).values
scaler=StandardScaler()

y=data["quality"].values
labelencoder =LabelEncoder()
y= labelencoder.fit_transform(y)

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)

scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
neighbors = np.arange(1, 60)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

for i, k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train,y_train)
    train_accuracy[i] = knn.score(x_train, y_train)
    test_accuracy[i] = knn.score(x_test, y_test)

plt.title('Accuracy for K values')
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
error_rate = []

for i in range(1,60):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    pred = knn.predict(x_test)
    error_rate.append(np.mean(pred != y_test))


plt.figure(figsize=(10,5))
plt.title('Error Rates for K values')
plt.plot(range(1,60),error_rate, marker='o', markersize=7)
plt.show()
```

<h4> A small value of K leads to overfitting and a large value of K leads to underfitting. From the accuracy and error graphs, K=11 is an optimal value. </h4>

```python
knn=KNeighborsClassifier(n_neighbors=11)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
print('K = 11')
print('Accuracy Score = ',accuracy_score(y_test,y_pred))
```

<h4> Using a modelling technique with removal of outliers and K value of 11, Accuracy = 0.6035 </h4>

<h3> KNN Modelling - 3 </h3>

<h4> In this model, we categorise 'Quality' based on Low : 3-4, Medium: 5-6, High: 7-8 into 'Category'. <br><br> Performing modelling on this data will give better results as the data becomes more suitable for a classification model. </h4> 

```python
cr_data=red_wines
cr_data.drop('free sulfur dioxide',axis=1)

quality = red_wines['quality'].values
category = []

for num in quality:
    if num < 5:
        category.append("Low")
    elif num > 6:
        category.append("High")
    else:
        category.append("Medium")
        
category = pd.DataFrame(data=category, columns=["category"])
cr_data = pd.concat([cr_data, category], axis=1)
cr_data.drop('quality', axis=1)

cr_X = cr_data.drop('category',axis=1)
cr_y = cr_data['category']

cr_X_train, cr_X_test, cr_y_train, cr_y_test = train_test_split(cr_X, cr_y, test_size = 0.20,random_state=42,stratify=cr_y)
```

```python
neighbors = np.arange(1, 60)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

for i, k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(cr_X_train,cr_y_train)
    train_accuracy[i] = knn.score(cr_X_train, cr_y_train)
    test_accuracy[i] = knn.score(cr_X_test, cr_y_test)

plt.title('Accuracy for K values')
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
error_rate = []

for i in range(1,60):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(cr_X_train, cr_y_train)
    pred = knn.predict(cr_X_test)
    error_rate.append(np.mean(pred != cr_y_test))

plt.figure(figsize=(10,5))
plt.title('Error Rates for K values')
plt.plot(range(1,60),error_rate, marker='o', markersize=7)
plt.show()
```

<h4> A small value of K leads to overfitting and a large value of K leads to underfitting. From the accuracy and error graphs, K=12 is an optimal value. </h4>

```python
knn = KNeighborsClassifier(n_neighbors=12)  
knn.fit(cr_X_train, cr_y_train)  
cr_y_pred = knn.predict(cr_X_test)  
print('K = 12')
print('Accuracy Score = ',accuracy_score(cr_y_test,cr_y_pred))
```

<h4> Using a modelling technique with recategorising and K value of 12, Accuracy = 0.859375 </h4>
